# Desarrollo de Modelos de Regresión

```python
# Seleccionar las variables relevantes basadas en el mapa de calor de correlaciones
# (se excluyen identificadores y columnas objetivo irrelevantes)
relevant_features = [
    'TCH', '%Sac.Caña', 'Area Neta', 'Edad Ult Cos', 'Rdto', 'TAH', '%ATR', '%Fibra Caña',
    '%ME Min', '%ME Veg', '%ME Tot', 'Pureza', 'Vejez', 'Num.Riegos', 'M3 Riego', 'DDUlt.Riego',
    'Lluvias (2 Meses Ant.)', 'Lluvias Ciclo', 'Lluvias 0 -3', 'Lluvias tres a seis',
    'Lluvias seis a nueve', 'Luvias 9 -FC', '%Infest.Diatrea', 'Fosfato Jugo', 'Urea 46%', 'MEZ',
    'Boro Granul.', 'MicroZinc', 'NITO_XTEND', 'Sul.Amonio', 'NITRAX-S', 'Vinaza',
    'Temp. Media 0-3', 'Temp. Media Ciclo', 'Temp Max Ciclo', 'Temp Min Ciclo',
    'Humedad Rel Media 0-3 ', 'Humedad Rel Media Ciclo',
    'Oscilacion Temp Med 0-3', 'Oscilacion Temp Ciclo',
    'Radicion Solar 0-3', 'Radiacion Solar Ciclo',
    'Precipitacion 0_3', 'Precipitacion Ciclo',
    'Evaporacion 0-3', 'Evaporacion Ciclo'
]

# Crear un DataFrame solo con esas columnas relevantes
df_model = df_historico[relevant_features].copy()

# -------- LIMPIEZA DE DATOS --------
# Eliminar filas con valores faltantes (NaN)
df_model.dropna(inplace=True)

# -------- DETECCIÓN Y ELIMINACIÓN DE OUTLIERS --------
# Método del rango intercuartílico (IQR)
Q1 = df_model.quantile(0.25)   # Primer cuartil (25%)
Q3 = df_model.quantile(0.75)   # Tercer cuartil (75%)
IQR = Q3 - Q1                  # Rango intercuartílico

# Definir límites inferior y superior (1.5 * IQR es el criterio estándar)
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Crear una máscara booleana que marque las filas con outliers
outliers_mask = ((df_model < lower_bound) | (df_model > upper_bound)).any(axis=1)

# Filtrar los datos eliminando los outliers
df_model_cleaned = df_model[~outliers_mask].copy()

# -------- INFORME DE LA LIMPIEZA --------
print(f"Número original de filas: {len(df_historico)}")
print(f"Número de filas después de eliminar valores faltantes: {len(df_model)}")
print(f"Número de filas después de eliminar outliers: {len(df_model_cleaned)}")

# Mostrar las primeras filas del DataFrame limpio
display(df_model_cleaned.head())
```

```python
# Calcular el porcentaje de valores faltantes por columna
missing_values_percentage = df_historico[relevant_features].isnull().sum() / len(df_historico) * 100

# Identificar columnas con más de 50% de valores faltantes
# (estas columnas se eliminan porque aportarían muy poca información al modelo)
columns_to_drop = missing_values_percentage[missing_values_percentage > 50].index.tolist()

# Crear un nuevo DataFrame eliminando esas columnas con demasiados nulos
df_model = df_historico[relevant_features].drop(columns=columns_to_drop).copy()

# Eliminar cualquier fila que aún tenga valores faltantes
df_model.dropna(inplace=True)

# -------- DETECCIÓN Y ELIMINACIÓN DE OUTLIERS --------
# Usamos el método del rango intercuartílico (IQR)
Q1 = df_model.quantile(0.25)   # Primer cuartil (25%)
Q3 = df_model.quantile(0.75)   # Tercer cuartil (75%)
IQR = Q3 - Q1                  # Rango intercuartílico

# Definir límites inferior y superior para detectar outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Crear una máscara que marca las filas que contienen algún valor fuera de los límites
outliers_mask = ((df_model < lower_bound) | (df_model > upper_bound)).any(axis=1)

# Eliminar los outliers del dataset
df_model_cleaned = df_model[~outliers_mask].copy()

# -------- INFORME DE LIMPIEZA --------
print(f"Número original de filas: {len(df_historico)}")
print(f"Número de filas después de eliminar columnas con muchos nulos y filas con nulos: {len(df_model)}")
print(f"Número de filas después de eliminar outliers: {len(df_model_cleaned)}")

# Mostrar las primeras filas del DataFrame limpio
display(df_model_cleaned.head())
```

```python
from sklearn.model_selection import train_test_split

# Separamos variables predictoras (X) y variables objetivo (y)
# Quitamos las columnas 'TCH' y '%Sac.Caña' de X porque son las variables que queremos predecir
X = df_model_cleaned.drop(['TCH', '%Sac.Caña'], axis=1)

# Definimos los targets (variables a predecir) para cada modelo
y_tch = df_model_cleaned['TCH']            # Para el modelo de Toneladas de caña por hectárea
y_sac_cana = df_model_cleaned['%Sac.Caña'] # Para el modelo de % de Sacarosa

# -------- DIVISIÓN DE DATOS PARA EL MODELO DE TCH --------
# Dividimos en entrenamiento (80%) y prueba (20%) con semilla aleatoria fija (random_state=42)
X_train_tch, X_test_tch, y_train_tch, y_test_tch = train_test_split(
    X, y_tch, test_size=0.2, random_state=42
)

# -------- DIVISIÓN DE DATOS PARA EL MODELO DE %Sac.Caña --------
X_train_sac_cana, X_test_sac_cana, y_train_sac_cana, y_test_sac_cana = train_test_split(
    X, y_sac_cana, test_size=0.2, random_state=42
)

# --------- REPORTE DE DIMENSIONES ---------
# Mostramos las dimensiones (número de filas y columnas) de cada conjunto
print("División de datos para el modelo TCH:")
print(f"  Conjunto de entrenamiento (features): {X_train_tch.shape}")
print(f"  Conjunto de prueba (features): {X_test_tch.shape}")
print(f"  Conjunto de entrenamiento (target): {y_train_tch.shape}")
print(f"  Conjunto de prueba (target): {y_test_tch.shape}")

print("\nDivisión de datos para el modelo %Sac.Caña:")
print(f"  Conjunto de entrenamiento (features): {X_train_sac_cana.shape}")
print(f"  Conjunto de prueba (features): {X_test_sac_cana.shape}")
print(f"  Conjunto de entrenamiento (target): {y_train_sac_cana.shape}")
print(f"  Conjunto de prueba (target): {y_test_sac_cana.shape}")
```

```python
from sklearn.linear_model import LinearRegression

# -------- MODELO DE REGRESIÓN PARA TCH --------
# Creamos una instancia del modelo de regresión lineal
model_tch = LinearRegression()

# Entrenamos el modelo usando los datos de entrenamiento (X_train_tch y y_train_tch)
model_tch.fit(X_train_tch, y_train_tch)


# -------- MODELO DE REGRESIÓN PARA %Sac.Caña --------
model_sac_cana = LinearRegression()
model_sac_cana.fit(X_train_sac_cana, y_train_sac_cana)


# Mensajes de confirmación
print("Modelo de regresión para TCH entrenado correctamente.")
print("Modelo de regresión para %Sac.Caña entrenado correctamente.")
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# ----------- EVALUACIÓN DEL MODELO PARA TCH -----------
# Generamos predicciones sobre los datos de prueba
y_pred_tch = model_tch.predict(X_test_tch)

# Calculamos el Error Cuadrático Medio (MSE) -> penaliza más los errores grandes
mse_tch = mean_squared_error(y_test_tch, y_pred_tch)

# Calculamos el Coeficiente de Determinación (R²) -> proporción de varianza explicada
r2_tch = r2_score(y_test_tch, y_pred_tch)

# Mostramos resultados para el modelo TCH
print(f"TCH Model - Mean Squared Error: {mse_tch:.2f}")
print(f"TCH Model - R-squared: {r2_tch:.2f}")


# ----------- EVALUACIÓN DEL MODELO PARA %Sac.Caña -----------
# Generamos predicciones sobre los datos de prueba
y_pred_sac_cana = model_sac_cana.predict(X_test_sac_cana)

# Calculamos métricas de error y bondad de ajuste
mse_sac_cana = mean_squared_error(y_test_sac_cana, y_pred_sac_cana)
r2_sac_cana = r2_score(y_test_sac_cana, y_pred_sac_cana)

# Mostramos resultados para el modelo %Sac.Caña
print(f"%Sac.Caña Model - Mean Squared Error: {mse_sac_cana:.2f}")
print(f"%Sac.Caña Model - R-squared: {r2_sac_cana:.2f}")
```

```python
print("Resumen del Rendimiento de los Modelos:")   # Título del resumen
print("-" * 40)                                   # Línea separadora para mayor claridad

# --------- RESUMEN DEL MODELO TCH ---------
print("Modelo TCH:")
print(f"  Error Cuadrático Medio (MSE): {mse_tch:.2f}")   # MSE del modelo TCH
print(f"  Coeficiente de Determinación (R²): {r2_tch:.2f}")   # R² del modelo TCH

print("\nInterpretación:")
print(f"El modelo TCH tiene un MSE bajo ({mse_tch:.2f}), lo que indica que los valores predichos de TCH están, en promedio, cercanos a los valores reales.")
print(f"El valor de R² = {r2_tch:.2f} significa que aproximadamente el {r2_tch*100:.0f}% de la variabilidad en TCH es explicada por las variables independientes del modelo.")

# --------- RESUMEN DEL MODELO %Sac.Caña ---------
print("\nModelo %Sac.Caña:")
print(f"  Error Cuadrático Medio (MSE): {mse_sac_cana:.2f}")   # MSE del modelo %Sac.Caña
print(f"  Coeficiente de Determinación (R²): {r2_sac_cana:.2f}")   # R² del modelo %Sac.Caña

print("\nInterpretación:")
print(f"El modelo %Sac.Caña también presenta un MSE muy bajo ({mse_sac_cana:.2f}), lo que sugiere predicciones muy precisas para esta variable.")
print(f"El valor de R² = {r2_sac_cana:.2f} indica que aproximadamente el {r2_sac_cana*100:.0f}% de la variabilidad en %Sac.Caña es explicada por las variables independientes del modelo.")

# --------- LISTADO DE VARIABLES ---------
print("\nVariables utilizadas en los modelos:")
print(X.columns.tolist())   # Imprime la lista de columnas usadas como predictores
```

```python
import statsmodels.api as sm

# Agregar un término constante a los conjuntos de características
# Esto corresponde al intercepto (β0) de la regresión lineal
X_train_tch_const = sm.add_constant(X_train_tch)
X_train_sac_cana_const = sm.add_constant(X_train_sac_cana)

# Ajustar modelos OLS (Ordinary Least Squares - Mínimos Cuadrados Ordinarios)
# Se ajustan dos modelos: uno para predecir TCH y otro para %Sac.Caña
ols_model_tch = sm.OLS(y_train_tch, X_train_tch_const).fit()
ols_model_sac_cana = sm.OLS(y_train_sac_cana, X_train_sac_cana_const).fit()

# Imprimir resúmenes completos de cada modelo
print("Resumen del modelo TCH:")
print(ols_model_tch.summary())    # Aquí se obtiene un reporte estadístico completo del modelo TCH

print("\nResumen del Modelo %Sac.Caña:")
print(ols_model_sac_cana.summary())  # Lo mismo, pero para %Sac.Caña
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

# ================================================
# 1. Eliminar variables de lluvias altamente correlacionadas
# (estas variables estaban mostrando multicolinealidad en el heatmap)
# ================================================
X_train_tch_reduced = X_train_tch.drop(['Lluvias 0 -3', 'Lluvias tres a seis',
                                        'Lluvias seis a nueve', 'Luvias 9 -FC'], axis=1)
X_train_sac_cana_reduced = X_train_sac_cana.drop(['Lluvias 0 -3', 'Lluvias tres a seis',
                                                  'Lluvias seis a nueve', 'Luvias 9 -FC'], axis=1)

# ================================================
# 2. Agregar la constante (intercepto) para los modelos OLS
# ================================================
X_train_tch_const_reduced = sm.add_constant(X_train_tch_reduced)
X_train_sac_cana_const_reduced = sm.add_constant(X_train_sac_cana_reduced)

# ================================================
# 3. Calcular el Factor de Inflación de Varianza (VIF)
# El VIF mide cuánto se infla la varianza de un coeficiente
# debido a la multicolinealidad con otras variables.
# Fórmula: VIF = 1 / (1 - R²)
# Valores de referencia:
#   - VIF = 1 → sin correlación
#   - 1 < VIF < 5 → multicolinealidad baja/moderada
#   - VIF > 10 → multicolinealidad alta (problema)
# ================================================

# Para el modelo de TCH
vif_data_tch = pd.DataFrame()
vif_data_tch["Variable"] = X_train_tch_const_reduced.columns
vif_data_tch["VIF"] = [variance_inflation_factor(X_train_tch_const_reduced.values, i)
                       for i in range(X_train_tch_const_reduced.shape[1])]

print("Valores VIF para el modelo TCH:")
display(vif_data_tch)

# Para el modelo de %Sac.Caña
vif_data_sac_cana = pd.DataFrame()
vif_data_sac_cana["Variable"] = X_train_sac_cana_const_reduced.columns
vif_data_sac_cana["VIF"] = [variance_inflation_factor(X_train_sac_cana_const_reduced.values, i)
                             for i in range(X_train_sac_cana_const_reduced.shape[1])]

print("\nValores VIF para el modelo %Sac.Caña:")
display(vif_data_sac_cana)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd
import statsmodels.api as sm

# ================================================
# 1. Reducir las variables eliminando predictores de 'Lluvias'
#    Estas variables estaban altamente correlacionadas entre sí
#    y podían generar multicolinealidad.
# ================================================
X_train_tch_reduced = X_train_tch.drop(['Lluvias 0 -3', 'Lluvias tres a seis',
                                        'Lluvias seis a nueve', 'Luvias 9 -FC'], axis=1)
X_train_sac_cana_reduced = X_train_sac_cana.drop(['Lluvias 0 -3', 'Lluvias tres a seis',
                                                  'Lluvias seis a nueve', 'Luvias 9 -FC'], axis=1)

# ================================================
# 2. Agregar la constante (intercepto) a los modelos OLS
#    Esto es necesario para que la regresión tenga el término β0.
# ================================================
X_train_tch_const_reduced = sm.add_constant(X_train_tch_reduced)
X_train_sac_cana_const_reduced = sm.add_constant(X_train_sac_cana_reduced)

# ================================================
# 3. Calcular el VIF (Variance Inflation Factor)
#    - VIF mide cuánto se infla la varianza de un coeficiente
#      debido a la correlación con otras variables.
#    - Valores de referencia:
#        VIF = 1 → sin correlación
#        1 < VIF < 5 → correlación moderada
#        VIF > 10 → multicolinealidad fuerte (problema)
# ================================================

# VIF para el modelo de TCH
vif_data_tch = pd.DataFrame()
vif_data_tch["Variable"] = X_train_tch_const_reduced.columns
vif_data_tch["VIF"] = [
    variance_inflation_factor(X_train_tch_const_reduced.values, i)
    for i in range(X_train_tch_const_reduced.shape[1])
]

print("Valores VIF para el modelo TCH:")
display(vif_data_tch)

# VIF para el modelo de %Sac.Caña
vif_data_sac_cana = pd.DataFrame()
vif_data_sac_cana["Variable"] = X_train_sac_cana_const_reduced.columns
vif_data_sac_cana["VIF"] = [
    variance_inflation_factor(X_train_sac_cana_const_reduced.values, i)
    for i in range(X_train_sac_cana_const_reduced.shape[1])
]

print("\nValores VIF para el modelo %Sac.Caña:")
display(vif_data_sac_cana)

# ================================================
# 4. Ajustar los modelos OLS nuevamente
#    Ahora con el conjunto de variables reducidas,
#    después de mitigar la multicolinealidad.
# ================================================
ols_model_tch_reduced = sm.OLS(y_train_tch, X_train_tch_const_reduced).fit()
ols_model_sac_cana_reduced = sm.OLS(y_train_sac_cana, X_train_sac_cana_const_reduced).fit()

# Mostrar resúmenes de los nuevos modelos
print("\nResumen del modelo TCH (con variables reducidas):")
print(ols_model_tch_reduced.summary())

print("\nResumen del modelo %Sac.Caña (con variables reducidas):")
print(ols_model_sac_cana_reduced.summary())
```
